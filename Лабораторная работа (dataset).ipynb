   Прислать python notebook с кодом и перыми 30 строками из получившегося датасета и примером трех получившихся бронирований.

```python
df = pd.read_excel('dataset.xlsx')
```

```python
df.head() 
```

```python
#Оставить один год.
df = df[df["arrival_year"] == 2018]
```

```python
df.head()
```

```python
#arrival_year, arrival_month, arrival_date объединить в один столбец (arrival_datetime) формата ГГГГ-ММ-ДД.
df.loc[(df['arrival_month'] == 4) & (df['arrival_date'] > 30), 'arrival_date'] = 30
```

```python
df['arrival_datetime'] = pd.to_datetime(
    df['arrival_year'].astype(str) + '-' + 
    df['arrival_month'].astype(str) + '-' + 
    df['arrival_date'].astype(str),
    errors='coerce'
).dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
df.drop(columns=["arrival_year", "arrival_month", "arrival_date"])
```

```python
# Посчитать для каждого бронирования столбец с днем бронирования (ГГГГ-ММ-ДД).
df["arrival_datetime"] = pd.to_datetime(df["arrival_datetime"], errors="coerce")

df["lead_time"] = pd.to_numeric(df["lead_time"], errors="coerce").fillna(0).astype(int)

df["booking_date"] = df["arrival_datetime"] - pd.to_timedelta(df["lead_time"], unit="D")

df["booking_date"] = df["booking_date"].dt.strftime('%Y-%m-%d')
```

```python
df.head(10)
```

```python
# Создать столбец "event" (текст), в каждой ячейке которого написать "Отменено" или "Подтверждено" 
# в зависимости от значения booking_status.

df["event"] = df["booking_status"].map({"Canceled": "Отменено", "Not_Canceled": "Подтверждено"})
```

```python
df.head()
```

```python
# Скопировать каждую строку с бронированием. В "event" добавить "Заезд". 
# Скопировать строку с каждым бронированием столько раз, сколько специальных запросов, 
# сделанных клиентом, по нему было (no_of_special_requests). 
# В каждой из скопированных строк заполнить "event" = "Дополнительный запрос".

# пустой список для хранения обработанных строк
new_rows = []

for _, row in df.iterrows():
    # запись "Заезд" (одно на бронирование)
    checkin_row = row.to_dict()
    checkin_row['event'] = 'Заезд'
    new_rows.append(checkin_row)
    
    # записи "Дополнительный запрос" (по количеству запросов)
    for _ in range(row['no_of_special_requests']):
        request_row = row.to_dict()
        request_row['event'] = 'Дополнительный запрос'
        new_rows.append(request_row)

# Создаем новый DataFrame
result_df = pd.DataFrame(new_rows)
```

```python
print(result_df[['Booking_ID', 'no_of_special_requests', 'event']].head(20))
```

```python
df.head()
```

```python
# Создать столбец "date", в который при "event" = "Отменено" или "Подтверждено" 
# добавить дату бронирования (ГГГГ-ММ-ДД), а при "event" = "Заезд" добавить arrival_datetime, 
# при "event" = "Дополнительный запрос" добавить случайную дату и время (ГГГГ-ММ-ДД) в промежутке
# от даты бронирования до даты заезда.

import numpy as np
from datetime import datetime, timedelta

df['arrival_datetime'] = pd.to_datetime(df['arrival_datetime'], errors='coerce')
df['booking_date'] = pd.to_datetime(df['booking_date'], errors='coerce')

# генерации случайной даты
def generate_random_date(row):
    if pd.isna(row['booking_date']) or pd.isna(row['arrival_datetime']):
        return np.nan
    
    delta = (row['arrival_datetime'] - row['booking_date']).days
    if delta <= 0:  # booking_date позже arrival_datetime
        return np.nan
    
    random_days = np.random.randint(0, delta + 1)
    random_date = row['booking_date'] + timedelta(days=random_days)
    return random_date.strftime('%Y-%m-%d')

def create_date_column(row):
    if row['event'] in ['Отменено', 'Подтверждено']:
        return row['booking_date'].strftime('%Y-%m-%d') if not pd.isna(row['booking_date']) else np.nan
    elif row['event'] == 'Заезд':
        return row['arrival_datetime'].strftime('%Y-%m-%d') if not pd.isna(row['arrival_datetime']) else np.nan
    elif row['event'] == 'Дополнительный запрос':
        return generate_random_date(row)
    else:
        return np.nan

result_df['date'] = result_df.apply(create_date_column, axis=1)
result_df['date'] = result_df['date'].replace({np.nan: ''})
```
